# **Extracting data from nested tags**

```python
div_notes = soup.find_all('div', {'role' : 'note'})
div_notes
```

```python
div_notes[0].find('a')
```

```python
div_notes[6].find('a')
```

```python
#create links list from div_notes
div_links = []
for div in div_notes:
    anchors = div.find_all('a')
    
    div_links.extend(anchors)
div_links
```

```python
div_urls = [urljoin(base_url, a.get('href')) for a in div_links if a.get('href') != 'None']
div_urls
```

# **Scraping multipages automatically**

```python
div_urls
```

```python
paragraph_text = []

i = 0
for url in div_urls:
    link = requests.get(url)
    
    if link.status_code == 200:
        print("url# {0}: {1}".format(i+1, url))
    else:
      print("Status code{0}: Skipping url# {1} : {2}".format(link.status_code, i+1, url))
      i = i + 1
      continue
    html_response = link.content
    soup_object = BeautifulSoup(html_response, 'lxml')
    note_paragraph  = soup_object.find_all('p')
    text_in_paragraph = [p.text for p in note_paragraph]

    paragraph_text.append(text_in_paragraph)

    i  = i + 1
paragraph_text
```

```python
paragraph_text[0]
```

```python
page_text = [" ".join(text) for text in paragraph_text]
page_text[9]
```

```python
url_to_text = dict(zip(div_urls, page_text))
```

```python
print(url_to_text['https://en.wikipedia.org/wiki/Elements_of_music'])
```
